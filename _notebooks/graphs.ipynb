```python
import pandas as pd
import matplotlib
```

```python
spreadsheet = pd.read_excel(r"C:\Users\fredd\Downloads\census.xls")
```

```python
spreadsheet.drop(columns={'Unnamed: 2', 'Unnamed: 3'}, inplace=True)
```

```python
countries = spreadsheet[~spreadsheet['Area name'].isna()]
countries
```

```python
countries.plot.bar(x="Area name", y="Males", title="line")
```

```python
main_countries = countries[countries['Area name'].isin(['WALES', 'SCOTLAND', 'ENGLAND', 'NORTHERN IRELAND'])]
main_countries.drop_duplicates(subset={'Area name'}, inplace=True)
main_countries
```

```python
colors = ['c', 'red', 'blue', 'green']
labels = ['England', 'Wales', 'Scotland', 'Northern Ireland']
main_countries.plot.pie(y="Males", title="pie graph of the amount of males of great britain", labels=labels, colors=colors)
```

```python
males_females_colors = ['ghostwhite', 'red']
main_countries.plot.bar(x='Area name', y=(['Males', 'Females']), color =males_females_colors)
```

```python
colors = ['c', 'red', 'blue', 'green']
labels = ['England', 'Wales', 'Scotland', 'Northern Ireland']
main_countries.plot.pie(y="All", title="pie graph of the amount of humans of great britain", labels=labels, colors=colors)
```
